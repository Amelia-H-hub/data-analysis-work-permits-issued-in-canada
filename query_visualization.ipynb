### Yearly Trend of Top 10 Occupations

```python
cur.execute("""
    SELECT w.en_year, n.class_title, SUM(w.total) AS total_sum
    FROM work_permit_holders AS w
    LEFT JOIN national_occupational_classifications AS n
    ON w.en_occupation = n.code
    WHERE n.class_title IN (
    	SELECT n.class_title
    	FROM work_permit_holders as w
    	LEFT JOIN national_occupational_classifications as n
    	ON w.en_occupation = n.code
    	WHERE w.en_occupation <> 'Othe' AND w.en_year <> 2025
    	GROUP BY w.en_occupation, n.class_title
    	ORDER BY SUM(w.total) DESC
    	LIMIT 10
    )
    GROUP BY w.en_year, n.class_title
    ORDER BY w.en_year
""")
rows_year_trend_top10_occup = cur.fetchall()
```

### Top 5 Occupations with Large Year-to-Year Variations

```python
cur.execute("""
    WITH top5_dramatic_occupation_year AS (
    	WITH year_sum AS (
    		SELECT w.en_year, n.class_title, SUM(w.total) AS total_sum
    		FROM work_permit_holders AS w
    		LEFT JOIN national_occupational_classifications AS n
    		ON w.en_occupation = n.code
    		WHERE w.en_year <> 2025
    		GROUP BY w.en_year, n.class_title
    		ORDER BY w.en_year
    	),
    	diffs_year AS (
    		SELECT class_title, en_year, total_sum,
    			LAG(total_sum) OVER (PARTITION BY class_title ORDER BY en_year) AS prev_total,
    			total_sum - LAG(total_sum) OVER (PARTITION BY class_title ORDER BY en_year) AS diff,
    			ROUND(
    				100.0 * (total_sum - LAG(total_sum) OVER (PARTITION BY class_title ORDER BY en_year)) / NULLIF(LAG(total_sum) OVER (PARTITION BY class_title ORDER BY en_year), 0),
    				1
    			) AS percent_change
    		FROM year_sum
    	),
    	diffs_year_ranked AS (
    		SELECT *,
    			ROW_NUMBER() OVER (PARTITION BY class_title ORDER BY ABS(percent_change) DESC) AS rn
    		FROM diffs_year
    		WHERE percent_change IS NOT NULL
    	)
    	SELECT *
    	FROM diffs_year_ranked
    	WHERE rn = 1
    	ORDER BY ABS(percent_change) DESC
    	LIMIT 5
    )
    SELECT w.en_year, n.class_title, SUM(w.total) AS total_sum
    FROM work_permit_holders AS w
    LEFT JOIN national_occupational_classifications AS n
    ON w.en_occupation = n.code
    WHERE n.class_title IN (SELECT class_title FROM top5_dramatic_occupation_year) AND w.en_year <> 2025
    GROUP BY w.en_year, n.class_title;
""")
rows_top5_occupation_dramatic_year_trend = cur.fetchall()
```

### Highlights: 5 Standout Yearly Trends Among Top 10 Jobs

### Quarterly Trend

```python
cur.execute("""
    SELECT en_quarter, SUM(total) AS total_sum
    FROM work_permit_holders
    WHERE en_year <> 2025
    GROUP BY en_quarter
    ORDER BY en_quarter
""")
rows_quarter_trend = cur.fetchall()
```

## Top 5 Occupations Showing Sharp Quarter-to-Quarter Changes

```python
cur.execute("""
    WITH top5_dramatic_occupation_quarter AS (
    	WITH quarter_sum AS (
    		SELECT w.en_quarter, n.class_title, SUM(w.total) AS total_sum
    		FROM work_permit_holders AS w
    		LEFT JOIN national_occupational_classifications AS n
    		ON w.en_occupation = n.code
    		WHERE w.en_year <> 2025
    		GROUP BY w.en_quarter, n.class_title
    		ORDER BY w.en_quarter
    	),
    	diffs AS (
    		SELECT class_title, en_quarter, total_sum,
    			LAG(total_sum) OVER (PARTITION BY class_title ORDER BY en_quarter) AS prev_total,
    			total_sum - LAG(total_sum) OVER (PARTITION BY class_title ORDER BY en_quarter) AS diff,
    			ROUND(
    				100.0 * (total_sum - LAG(total_sum) OVER (PARTITION BY class_title ORDER BY en_quarter)) / NULLIF(LAG(total_sum) OVER (PARTITION BY class_title ORDER BY en_quarter), 0),
    				1
    			) AS percent_change
    		FROM quarter_sum
    	)
    	SELECT *
    	FROM diffs
    	WHERE percent_change IS NOT NULL
    	ORDER BY ABS(percent_change) DESC
    	LIMIT 5
    )
    ---- Step 2: Select Quarter Trend from Top 5 Dramatic Occupations
    SELECT w.en_quarter, n.class_title, SUM(w.total) AS total_sum
    FROM work_permit_holders AS w
    LEFT JOIN national_occupational_classifications AS n
    ON w.en_occupation = n.code
    WHERE n.class_title IN (SELECT class_title FROM top5_dramatic_occupation_quarter) AND w.en_year <> 2025
    GROUP BY w.en_quarter, n.class_title;
""")
rows_top5_occupations_dramatic_quarter_trend = cur.fetchall()
```

### Top 10 Occupations by Number of Work Permits Issued in Canada

```python
cur.execute("""
    SELECT w.en_occupation, n.class_title, SUM(w.total) as total_sum
    FROM work_permit_holders as w
    LEFT JOIN national_occupational_classifications as n
    ON w.en_occupation = n.code
    WHERE w.en_occupation <> 'Othe' AND w.en_year <> 2025
    GROUP BY w.en_occupation, n.class_title
    ORDER BY total_sum DESC
    LIMIT 10
""")
rows_top10_occup = cur.fetchall()
```

### Top 10 Occupations in BC

```python
cur.execute("""
    SELECT w.en_occupation, n.class_title, SUM(w.total) as total_sum
    FROM work_permit_holders as w
    LEFT JOIN national_occupational_classifications as n
    ON w.en_occupation = n.code
    WHERE w.en_province_territory = 'British Columbia' AND w.en_occupation <> 'Othe' AND w.en_year <> 2025
    GROUP BY w.en_occupation, n.class_title
    ORDER BY total_sum DESC
    LIMIT 10
""")
rows_top10_occup_bc = cur.fetchall()
```

### Total Work Permits Issued by Province

```python
cur.execute("""
    SELECT en_province_territory, SUM(total) as total_sum
    FROM work_permit_holders
    WHERE en_province_territory <> 'Province/territory not stated'
    GROUP BY en_province_territory
    ORDER BY en_province_territory
""")
rows_total_province = cur.fetchall()
```

### Top Occupacation in Each Province

```python
cur.execute("""
    SELECT pr.en_province_territory, pr.iso_code, n.class_title, pr.total
    FROM(
    	SELECT w.en_province_territory, c.iso_code, w.en_occupation, SUM(w.total) as total,
    		ROW_NUMBER() OVER (PARTITION BY w.en_province_territory ORDER BY SUM(w.total) DESC) AS rn
    	FROM work_permit_holders as w
    	LEFT JOIN canada_regions as c
    	ON w.en_province_territory = c.province
    	WHERE w.en_occupation <> 'Othe' AND c.iso_code IS NOT NULL
    	GROUP BY w.en_province_territory, w.en_occupation, c.iso_code
    ) AS pr
    LEFT JOIN national_occupational_classifications as n
    ON pr.en_occupation = n.code
    WHERE pr.rn = 1
    ORDER BY pr.en_province_territory
""")
rows_top_occupation_each_province = cur.fetchall()
```

### Total Work Permits Issued by Province

```python
df_total_province = pd.DataFrame(rows_total_province, columns=["province", "total"])

with open("data/georef-canada-province-fixed.geojson", "r") as f:
    canada_geojson_data = json.load(f)

# Add total to geojson properties
## Create a province -> total dict 
province_total_dict = dict(zip(df_total_province['province'], df_total_province['total']))
## Add total into properties of canada_geojson_data's features
for feature in canada_geojson_data['features']:
    prov_name = feature['properties']['prov_name_en']
    feature['properties']['total'] = province_total_dict.get(prov_name, 0)

# Create Folium map (set center to canada)
m = folium.Map(location=[56.1304, -106.3468], zoom_start=3)

# Add Choropleth
folium.Choropleth(
    geo_data=canada_geojson_data,
    name='choropleth',
    data=df_total_province,
    columns=['province', 'total'],
    key_on='feature.properties.prov_name_en',
    fill_color='YlGnBu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Total Work Permits issued by Province'
).add_to(m)

# Add Tooltip
folium.GeoJson(
    canada_geojson_data,
    name="Province Names",
    style_function=lambda feature: {
        'fillColor': feature['properties'].get('color', 'gray'),
        'color': 'black',
        'weight': 1,
        'fillOpacity': 0
    },
    tooltip=folium.GeoJsonTooltip(
        fields=["prov_name_en", "total"],
        aliases=["Province", "Total Work Permits"],
        localize=True,
        sticky=False
    )
).add_to(m)

# Save as HTML
m.save('total_work_permits_issued_by_province.html')
```

### Top Occupacation in Each Province

```python
df_top_occupation_each_province = pd.DataFrame(rows_top_occupation_each_province, columns=["province", "iso_code", "occupation", "total"])

with open("data/georef-canada-province-fixed.geojson", "r") as f:
    canada_geojson_data = json.load(f)

occupation_color_map = {
    "Food counter attendants, kitchen helpers and related support occupations": "#FFC1CC",
    "Food service supervisors": "#FFB6C1",
    "Nurse aides, orderlies and patient service associates": "#D5F0A3",
    "University professors and lecturers": "#F9D390",
    "Software engineers and designers": "#BFD7FF",
    "Information systems analysts and consultants": "#C8E1FF",
    "Industrial butchers and meat cutters, poultry preparers and related workers": "#FF9999",
    "Air pilots, flight engineers and flying instructors": "#B0B0B0",
    "Water transport deck and engine room crew": "#7D99A7",
    "Other religious occupations": "#FDEBD0"
}

for feature in canada_geojson_data["features"]:
    province_name = feature["properties"]["prov_name_en"]
    matched_row = df_top_occupation_each_province[df_top_occupation_each_province["province"] == province_name]
    if not matched_row.empty:
        occ = matched_row.iloc[0]["occupation"]
        feature["properties"]["top_occupation"] = occ
        feature["properties"]["color"] = occupation_color_map.get(occ, '#cccccc')

# Create Folium map (set center to canada)
m = folium.Map(location=[56.1304, -106.3468], zoom_start=3)

folium.GeoJson(
    canada_geojson_data,
    style_function=lambda feature: {
        'fillColor': feature['properties'].get('color', 'gray'),
        'color': 'black',
        'weight': 1,
        'fillOpacity': 0.8
    },
    tooltip=folium.GeoJsonTooltip(
        fields=["prov_name_en", "top_occupation"],
        aliases=["Province", "Top Occupation"],
        localize=True
    )
).add_to(m)

# Save as HTML
m.save('top_occupation_province.html')
```
